```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tau200=[16.700000000000003, 16.849999999999998, 15.790000000000003]
tau10=[11.760000000000005, 10.579999999999998, 2.1500000000000008]
tau20=[13.920000000000002, 13.260000000000003, 6.7599999999999971]
```

```python
taus=[10,20,200]
h=[tau10[0],tau20[0],tau200[0]]
a=[tau10[1],tau20[1],tau200[1]]
r=[tau10[2],tau20[2],tau200[2]]
```

```python
plt.plot(taus,h,'k')
plt.plot(taus,a,'k',linestyle='-.')
plt.plot(taus,r,'k',linestyle='--')
plt.legend(('Hybrid', 'Amplitude', 'Rate'))
plt.xlabel('Membrane Constant (ms)')
plt.ylabel('Number of Spikes')
axes = plt.gca()
axes.set_xlim([0,250])
axes.set_ylim([0,20])
```

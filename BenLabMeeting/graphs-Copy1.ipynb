```python
hMean=np.zeros(20)
aMean=np.zeros(20)
rMean=np.zeros(20)
hTimeMean=np.zeros(20)
hCount=np.zeros(20)
aTimeMean=np.zeros(20)
aCount=np.zeros(20)
rTimeMean=np.zeros(20)
rCount=np.zeros(20)
for i in range(0,20):
    for j in range(0,100):
        hMean[i]=hMean[i]+len(hybridSpikeMat[i,j])/100
        aMean[i]=aMean[i]+len(ampSpikeMat[i,j])/100
        rMean[i]=rMean[i]+len(rateSpikeMat[i,j])/100
        if len(hybridSpikeMat[i,j]>0):
            hTimeMean[i]=hTimeMean[i]+np.mean(hybridSpikeMat[i,j]%.2)
            hCount[i]=hCount[i]+len(hybridSpikeMat[i,j])
            print(np.var(hybridSpikeMat[i,j]%.2))
        if len(ampSpikeMat[i,j]>0):
            aTimeMean[i]=aTimeMean[i]+np.mean(ampSpikeMat[i,j]%.2)
            aCount[i]=aCount[i]+len(ampSpikeMat[i,j])
        if len(rateSpikeMat[i,j]>0):
            rTimeMean[i]=rTimeMean[i]+np.mean((rateSpikeMat[i,j]%.2))
            rCount[i]=rCount[i]+len(rateSpikeMat[i,j])
    hTimeMean[i]=hTimeMean[i]/hCount[i]
    aTimeMean[i]=aTimeMean[i]/aCount[i]
    rTimeMean[i]=rTimeMean[i]/rCount[i]
```

```python
plt.plot(means[:,0],hMean,'k')
plt.plot(means[:,1],aMean,'k',linestyle='-.')
plt.plot(means[:,2],rMean,'k',linestyle='--')
plt.legend(('Hybrid', 'Amplitude', 'Rate'))
plt.xlabel('umber of Input Vesicles')
plt.ylabel('Number of Spikes')
axes = plt.gca()
axes.set_xlim([0,65])
axes.set_ylim([0,20])
```
